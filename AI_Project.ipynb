```python
learning_rates = [0.01, 0.05, 0.1]
epochs = [50, 100]
results = {'Model': [], 'Learning Rate': [], 'Epochs': [], 'Accuracy': []}
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
models = {
    'LR': LogisticRegression(),
    'SVC': SVC(),
    'KNN': KNeighborsClassifier(),
    'DT': DecisionTreeClassifier(),
    'RF': RandomForestClassifier()
}
```

```python
for model_name, model in models.items():
    for lr in learning_rates:
        for epoch in epochs:
            # Set hyperparameters
            if model_name == 'LR':
                model.set_params(C=lr)
            elif model_name == 'SVC':
                model.set_params(C=lr)
            elif model_name == 'KNN':
                model.set_params(n_neighbors=epoch)
            elif model_name == 'DT':
                model.set_params(max_depth=epoch)
            elif model_name == 'RF':
                model.set_params(n_estimators=epoch)

            # Train the model
            model.fit(X_train, y_train)

            # Predict on test set
            y_pred = model.predict(X_test)

            # Calculate accuracy
            accuracy = accuracy_score(y_test, y_pred)

            # Store the results
            results['Model'].append(model_name)
            results['Learning Rate'].append(lr)
            results['Epochs'].append(epoch)
            results['Accuracy'].append(accuracy)
```

```python
results_df = pd.DataFrame(results)

highlighted_results_df = results_df.copy()
highlighted_results_df['Best'] = results_df.groupby('Model')['Accuracy'].transform('max') == results_df['Accuracy']

print(highlighted_results_df)
```
